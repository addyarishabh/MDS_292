### Question 2: 
    
Write a program using any object-oriented language to find Sum of series:

Sum = 1 + 1/2! + 2/3! + 3/4! + ..... n

```python
# the number of terms we wants to add. 
n=int(input('Enter the range: '))
s=1
# outer loop
for i in range(1,n):
    m=1
    # inner loop
    for j in range(1,i+2):
        m=m*j
    s=s+(i/m)
print(f"The sum of series {n} is: ",s)
```

### Question 3:

Write a program using any object-oriented language to show the pattern: 

a) 

      1
    
     121
 
    12321

```python
# Reading number of rows
row = int(input('Enter no. of lines: '))

# Generating pattern
for i in range(1,row+1):
    
    # for space
    for j in range(1, row+1-i):
        print(' ', end='')
    
    # for increasing pattern
    for j in range(1,i+1):
        print(j, end='')
    
    # for decreasing pattern 
    for j in range(i-1,0,-1):
        print(j, end='')
    
    # Moving to next line
    print()
```

### Question 5:

Write a program to find "Armstring Number" and "Disarium Number" using any object-oriented language. 

##### Armstrong Number:  (Such as: 153, 370, 371 etc.)



```python
# take number as an user input
n=int(input('Enter the number: '))
s=0
m=n
while(n>0):
    rem=n%10
    s=s+(rem**3)
    n=n//10
if(s==m):
    print(f"The number {m} is an armstrong number.")
else:
    print(f"The number {m} is not an armstrong number.")
```

##### Disarium Numner:

A number is called Disarium if sum of its digits powered with their respective positions is equal to the number itself.

For example; number= 135

1^1 + 3^2 + 5^3 = 135

Therefore, 135 is a Disarium number.

```python
n = input('Enter the number: ')
# Count digits in n.
i= len(n)
# conert number from string to integer
n=int(n)
# copy the number to the another variable
m=n
result = 0
while(n!=0):
    digit = n%10
    result=result+pow(digit,i)
    n=int(n/10)
    i = i - 1
if(result==m):
    print(f"The number {m} is a Disarium Number!")
else:
    print(f"The number {m} is not an Disarium Number!")
```
